```python
n = len(data)
data = np.array(data)

for i in range(len(data.T)):
    data[:,i] = (data[:,i]-min(data[:,i]))/(max(data[:,i])-min(data[:,i]))

data = np.concatenate((np.ones(n).reshape(-1,1), data), axis =1)
savetxt('combined.csv', data, delimiter=',')
```
